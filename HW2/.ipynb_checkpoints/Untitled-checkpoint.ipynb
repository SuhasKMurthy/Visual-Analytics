```python
from bokeh.transform import jitter
from bokeh.models import ColumnDataSource, Legend
from bokeh.palettes import inferno

IDS = list(dfcomplete.id.unique())

source = ColumnDataSource(dfcomplete)

col = inferno(6)

p = figure(plot_width=800, plot_height=300, x_range=IDS, title="Product Spending")

c1 = p.circle(x=jitter('id', width=0.6, range=p.x_range), y='Fresh', color=col[0], source=source, alpha=0.3)
c2 = p.circle(x=jitter('id', width=0.6, range=p.x_range), y='Milk', color=col[1], source=source, alpha=0.3)
c3 = p.circle(x=jitter('id', width=0.6, range=p.x_range), y='Grocery', color=col[2], source=source, alpha=0.3)
c4 = p.circle(x=jitter('id', width=0.6, range=p.x_range), y='Frozen', color=col[3], source=source, alpha=0.3)
c5 = p.circle(x=jitter('id', width=0.6, range=p.x_range), y='Detergents_Paper', color=col[4], source=source, alpha=0.3)
c6 = p.circle(x=jitter('id', width=0.6, range=p.x_range), y='Delicassen', color=col[5], source=source, alpha=0.3)

legend = Legend(items=[
    ("fresh"   , [c1]),
    ("milk" , [c2]),
    ("grocery" , [c3]),
    ("frozen" , [c3]),
    ("detergents_paper" , [c3]),
    ("delicassen" , [c3]),
], location=(0, -30))

legend.click_policy = "hide"

p.add_layout(legend, 'left')

show(p)
```
